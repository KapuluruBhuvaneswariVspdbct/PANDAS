```python
import pandas as pd 
df=pd.DataFrame({
    'city': [
        'Mumbai', 'Mumbai', 'Mumbai city', 'Chennai', 'Chennai',
        'Delhi', 'Delhi', 'Bangalore', 'Bangalore', 'Mumbai',
        'Chennai', 'Delhi', 'Mumbai city'
    ],
    'branch': [
        'A', 'B', 'C', 'A', 'B',
        'A', 'C', 'A', 'B', 'C',
        'C', 'B', 'A'
    ],
    'category': [
        'Electronics', 'Groceries', 'Electronics', 'Groceries', 'Clothing',
        'Clothing', 'Electronics', 'Groceries', 'Clothing', 'Clothing',
        'Electronics', 'Groceries', 'Electronics'
    ],
    'sales': [
        20000, 15000, 18000, 12000, 8000,
        10000, 14000, 9000, 7000, 16000,
        13000, 9000, 21000
    ],
    'profit': [
        4000, 2000, 3000, 1000, 600,
        1200, 2800, 900, 400, 2300,
        1500, 800, 3500
    ]
})
df
```

```python
d=df.groupby('city')
d
```

```python
for city ,data in d:
    print(city)
    print()
    print(data)
    print()
```

```python
def foo(df,indx,col):
    if df[col].loc[indx]<=1500:
        return "1500"
    else:
        return "more"
k=df.groupby(lambda idx:foo(df,idx,'profit'))
k
```

```python
for c, g in k:
    print(c,g)
```
